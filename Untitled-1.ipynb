```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import scipy.stats as stats
```

```python
#Create 1d example

num_betas = 5
x = np.linspace(-0.5, 0.5, 1000)
beta_range = [5, 40]
mean_d = 0.5
means = np.random.uniform(-mean_d, mean_d, num_betas)
stds = np.random.uniform(mean_d/500, mean_d/50, num_betas)

Spiky_pdf = np.zeros_like(x)
dists = []
for i in range(num_betas):
    dist = stats.norm(loc=means[i], scale=stds[i])
    pdf = dist.pdf(x)
    pdf /= num_betas*np.trapezoid(pdf, x) #Normalize to make it a valid PDF
    Spiky_pdf += pdf
    dists.append(dist)

Flat_pdf = np.ones_like(x)/2

Bell_pdf = stats.norm.pdf(x, loc=0.0, scale=0.15)

print("Spiky PDF integral:", np.trapezoid(Spiky_pdf, x))
```

```python
#Sample high dimensional points from these distributions, plot random prior

Num_Dims = 27
Bulk_Samples = 10_000
Dist_Samples = 500

#Get samples
Bulk_Samples = np.random.uniform(-0.5, 0.5, (Bulk_Samples, Num_Dims))
Flat_Samples = np.random.uniform(-0.5, 0.5, (Dist_Samples, Num_Dims))
Bell_Samples = np.random.normal(0.0, 0.15, (Dist_Samples, Num_Dims))
Spiky_Samples = np.zeros((Dist_Samples, Num_Dims))

perplex = 30
iters = 1000

#Train T-SNE on Bulk & Flat Samples
tsne_1 = TSNE(n_components=2, perplexity=perplex, max_iter=iters, random_state=42)
Bulk_Data_1 = np.vstack((Bulk_Samples, Flat_Samples))
Bulk_TSNE_1 = tsne_1.fit_transform(Bulk_Data_1)
Bulk_Samples_1 = Bulk_TSNE_1[:Bulk_Samples.shape[0]]
Flat_Samples_1 = Bulk_TSNE_1[Bulk_Samples.shape[0]:]

#Train T-SNE on Flat & Bell Samples
tsne_2 = TSNE(n_components=2, perplexity=perplex, max_iter=iters, random_state=42)
Bulk_Data_2 = np.vstack((Bulk_Samples, Bell_Samples))
Bulk_TSNE_2 = tsne_2.fit_transform(Bulk_Data_2)
Bulk_Samples_2 = Bulk_TSNE_2[:Bulk_Samples.shape[0]]
Bell_Samples_2 = Bulk_TSNE_2[Bulk_Samples.shape[0]:]

#Train T-SNE on Flat & Spike Samples
#Generate a spiky dist for each dimension
means = np.random.uniform(-mean_d, mean_d, (num_betas, Num_Dims))
stds = np.random.uniform(mean_d/500, mean_d/50, (num_betas, Num_Dims))
dists = []
for i in range(num_betas):
    dist = stats.multivariate_normal(mean=means[i], cov=np.diag(stds[i]**2))
    dists.append(dist)

for i in range(Dist_Samples):
    idx = np.random.randint(0, num_betas, 1)[0]
    dist = dists[idx]
    Spiky_Samples[i, :] = dist.rvs()

    
print('jo')
tsne_3 = TSNE(n_components=2, perplexity=perplex, max_iter=iters, random_state=42)
Bulk_Data_3 = np.vstack((Bulk_Samples, Spiky_Samples))
Bulk_TSNE_3 = tsne_3.fit_transform(Bulk_Data_3)
Bulk_Samples_3 = Bulk_TSNE_3[:Bulk_Samples.shape[0]]
Spiky_Samples_3 = Bulk_TSNE_3[Bulk_Samples.shape[0]:]
```

```python
#Plot results
import matplotlib.gridspec as gs
fig = plt.figure(figsize=(12, 8))
#want pdf plot on top, all 3 tsne plots below
grid = gs.GridSpec(2, 3, figure=fig, height_ratios=[1, 1], width_ratios=[1, 1, 1], hspace=0.25)
#Create pdf plot across entire top row
axs = np.empty((2, 2), dtype=object)
axs[0, 0] = fig.add_subplot(grid[0, :])
axs[0, 1] = fig.add_subplot(grid[1, 0])
axs[1, 0] = fig.add_subplot(grid[1, 1])
axs[1, 1] = fig.add_subplot(grid[1, 2])

#Params
bulk_color = "#CCCCCC"
flat_color = "#424242"
bell_color = "#4545E4"
spiky_color = "#EC3D3D"


#Plot PDFs
axs[0, 0].plot(x, Flat_pdf, label='Flat PDF', color=flat_color, linestyle='--', linewidth=2)
axs[0, 0].plot(x, Bell_pdf, label='Bell PDF', color=bell_color, linestyle=':', linewidth=2)
axs[0, 0].plot(x, Spiky_pdf/10, label='Spiky PDF (0.1x)', color=spiky_color, linewidth=1)
#axs[0, 0].set_ylim(0, 2.5)

#Plot Flat
axs[0, 1].scatter(Bulk_Samples_1[:, 0], Bulk_Samples_1[:, 1], s=1, color=bulk_color, label='Bulk Samples')
axs[0, 1].scatter(Flat_Samples_1[:, 0], Flat_Samples_1[:, 1], s=10, color=flat_color, label='Flat Samples')

#Plot Bell
axs[1, 0].scatter(Bulk_Samples_2[:, 0], Bulk_Samples_2[:, 1], s=1, color=bulk_color, label='Bulk Samples')
axs[1, 0].scatter(Bell_Samples_2[:, 0], Bell_Samples_2[:, 1], s=10, color=bell_color, label='Bell Samples')

#Plot Spiky
axs[1, 1].scatter(Bulk_Samples_3[:, 0], Bulk_Samples_3[:, 1], s=1, color=bulk_color, label='Bulk Samples')
axs[1, 1].scatter(Spiky_Samples_3[:, 0], Spiky_Samples_3[:, 1], s=10, color=spiky_color, label='Spiky Samples')


#General Formatting
fig.set_facecolor('#F2F2F2')

#PDF Plot Formatting
axs[0, 0].set_title('Comparison of PDFs & T-SNE (27D)', fontsize='x-large')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('Probability Density')
axs[0, 0].legend(fontsize='large', frameon=False)

#Flat T-SNE Formatting

#Bell T-SNE Formatting

#Spiky T-SNE Formatting

#Place common legend to tsne plots above them
labels = ['Bulk Samples', 'Flat Samples', 'Bell Samples', 'Spiky Samples']
colors = [bulk_color, flat_color, bell_color, spiky_color]
handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=8) for color in colors]
axs[0, 1].legend(handles, labels, loc='upper center', bbox_to_anchor=(1.7, 1.15), ncol=4, fontsize='large', frameon=False)

#Common Formatting
for ax in axs.flat:
    if ax != axs[0, 0]:
        #Common Formatting to T-sne plots
        #ax.spines['top'].set_visible(False)
        #ax.spines['right'].set_visible(False)
        #ax.spines['left'].set_visible(False)
        #ax.spines['bottom'].set_visible(False)
        ax.set_xticks([])
        ax.set_yticks([])
    #ax.set_facecolor('#F2F2F2')
    
plt.savefig('TSNE_1D_Distributions.png', dpi=96, bbox_inches='tight')

#Find where it saved
import os
print("Saved figure to:", os.path.abspath('TSNE_1D_Distributions.svg'))
```
